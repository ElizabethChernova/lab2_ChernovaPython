5. (9б). Створіть клас Фігура без атрибутів, з методом get_area для отримання площі фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму площин фігур self та other.

```python
class Figure:
    def get_area(self):
        return 0

    def __add__(self, other):
        if isinstance(other, Figure):
            return self.get_area() + other.get_area()
        else:
            raise ValueError("Unsupported operand. The second operand must be of type Figure.")


figure1 = Figure()
figure2 = Figure()

area_sum = figure1 + figure2
print(f"Сума площ: {area_sum}")
```

```python


import math
class Triangle(Figure):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def get_area(self):
        return 0.5 * self.base * self.height

class Circle(Figure):
    def __init__(self, radius):
        self.radius = radius

    def get_area(self):
        return math.pi * self.radius**2


triangle = Triangle(base=5, height=8)
circle = Circle(radius=4)
triangle_area = triangle.get_area()
circle_area = circle.get_area()
total_area = triangle + circle

print(f"Triangle area: {triangle_area}")
print(f"Circle area: {circle_area}")
print(f"Total area (Triangle + Circle): {total_area}")
```

8. (13б). Створіть клас BankAccount з приватними атрибутами balance та account_number. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self._balance = initial_balance
        self._account_number = account_number

    @property
    def balance(self):
        return self._balance

    @property
    def account_number(self):
        return self._account_number

    @balance.setter
    def balance(self, amount):
        if amount < 0:
            print("Amount must be non-negative.")
        else:
            self._balance = amount
            print(f"Balance updated: {self._balance}")

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount}. New balance: {self._balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew {amount}. New balance: {self._balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")


account = BankAccount(account_number="12345", initial_balance=1000)
print(f"Account Number: {account.account_number}")
print(f"Initial Balance: {account.balance}")

account.deposit(500)
account.withdraw(200)

account.balance = 1500
print(f"Updated Balance: {account.balance}")
```

9. (11б). Створіть клас Library, який містить список об'єктів типу Book. Реалізуйте методи для додавання книги, видалення книги та відображення списку книг.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Added book: {book.title} by {book.author}")

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"Removed book: {book.title} by {book.author}")
        else:
            print("Book not found in the library.")

    def display_books(self):
        print("Library Books:")
        for book in self.books:
            print(f"{book.title} by {book.author}")


library = Library()

book1 = Book(title="Python Basics", author="John Doe")
book2 = Book(title="Web Development", author="Jane Smith")

library.add_book(book1)
library.add_book(book2)
library.display_books()

library.remove_book(book1)
library.display_books()
```

10. (13б). Створіть клас Person з атрибутами name та age. Створіть ще один клас Employee з такими атрибутами, як department та salary. Створіть клас Manager, який успадковує обидва класи Person та Employee. Продемонструйте використання множинної спадковості, створивши обʼєкт класу Manager та отримавши mro для цього класу.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Employee:
    def __init__(self, department, salary):
        self.department = department
        self.salary = salary

class Manager(Person, Employee):
    def __init__(self, name, age, department, salary):
        # Виклик конструкторів батьківських класів
        Person.__init__(self, name, age)
        Employee.__init__(self, department, salary)

manager = Manager(name="Lisa", age=20, department="IT", salary=80000)

print(f"Manager: {manager.name}, Age: {manager.age}, Department: {manager.department}, Salary: {manager.salary}")

print("MRO for Manager class:", Manager.mro())
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає літери англійської абетки, поки вони не вичерпаються.

```python
class AlphabetIterator:
    def __init__(self):
        self.current_letter = ord('a')

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_letter <= ord('z'):
            letter = chr(self.current_letter)
            self.current_letter += 1
            return letter
        else:
            raise StopIteration

alphabet_iterator = AlphabetIterator()


for _ in range(26):
    print(next(alphabet_iterator))
```
